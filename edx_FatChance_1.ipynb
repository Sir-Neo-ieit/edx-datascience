```python
#5.1 Lesson: The Monty Hall Game
# n doors, one of them has a car and n-1 doors goats
#prob of winning a car is 1/n, prob of loosing a car is (n-1)/n
#after opening 2 doors with goats, the remaining doors is n-2
#prob of winning if you switch 1/(n-2)
#original prob of loosing a car is (n-1)/n
#original prob of winning (AFTER switching) is (n-1)/(n-2) * (1/n)
#since (n-1)/(n-2) > 1 => original prob of winning (AFTER switching) is always > (1/n)
```

```python
#prob of event A occurs, P(A) = 1/n (prob of winning)
#prob of event B occurs, P(B) = (n-1)/n (prob of loosing)
# P(A) + P(B) = 1
# n doors with k cars
#P(A) = k/n and P(B) = (n-k)/n
#prob of winning if we DONT switch, k/n
#let p(W) be prob of Win
#P(W assuming A) => we assume that the first choise we made is a win
#P(w assumgin A) = (k-1) / (n-2)
#P(W and A) = P(w assumina A) * P(A) = (k-1) / (n-2) * (k/n)
#P(W assuming B) => we assume that the first choise we made is a loose
#P(W assuming B) = k / (n-2)
#P(W and B) = P(W assuming B) * P(B) = k/(n-2)  * (n-k)/n
#P(W) = P(A)*P(W assuming A) + P(B)*P(W assuming B)
#P(W) = k/n * (k-1)/(n-2)  + (n-k)/n * k/(n-2)
#P(W) = k/n * (n-1)/(n-2) , prob of winning with making a switch
#P(A), porb of winning without making a switch
#(n-1)/(n-2) > 1 => k/n * (n-1)/(n-2) > k/n => P(W) > P(A)
```

```python
#5.2 Lesson: Conditional Probability: Set-Up and Examples
#fair coin => prob H - 1/2, prob T - 1/2
#unfair coin or weighted coin => prob H - 1/4, prob of T - 3/4
# 3 coins flip, two fair coins and one weighted coin, play N times
# two fair coins pick up - 2/3 * N
# one weighted coin pick up - 1/3 * N
# in 2/3 * N games,  if one fiar coin picks up => 1/2 * 2/3 * N => 1/3 * N wins
# in 1/3 * N games, if weighted coin picks up and prob of H => 1/4 * 1/3 * N => 1/12 * N wins
# total of wins for player A => 1/3 * N + 1/12 * N => 5/12 * N wins
```

```python
# P(A) + P(B) = 1
# for events of N times => P(A) * N + P(B) * N
# P(W assuming A), the second event W where the first event was A
# total of events that result first event as A and second event as W => P(W assuming A) * P(A) * N
# P(W assuming B), the second event W where the first event was B
# total of events that result first event as B and second event as W => P(W assuming B) * P(B) * N
# P(W) - prob of the event W
# P(W) = P(W assuming A) * P(A) * N   +  P(W assuming B) * P(B) * N
```

```python
# if the first event A has k different results
# A1, A2, ..., Ak results
# P(Ai) - porb of (k)th result
# P(W assuming Ai) - prob of second event W where first event was Ai
# P(W) = P(W assuming A1) * P(A1) + P(W assuming A2) * P(A2) + ... + P(W assuming Ak) * P(Ak)
```

```python
# two players, player N and player C, rolls a dice each and the higher one wins.
# if is a tie, player N rolls a dice and (1,2,3,4) player N wins and (5,6) player C wins
# outcomes : player N wins    player C wins   Tie
# P(AN) - player N wins in the first event A
# P(AC) - player C wins in the first event A
# P(AT) - ties in the first event A
6*6 #possible outcomes
# 6 out of 36 times, it is a tie
# 15 out of 36 times, player N wins
# 15 out of 36 times, player C wins
# P(AN) = 15/36
# P(AC) = 15/36
# P(AT) = 6/36
# P(W assuming AN) - prob of player N wins if the first event was AN => 1
# P(W assuming AC) - prob of player N wins if the first event was AN => 0
# P(W assuming AT) - prob of player N wins if the first event was AT => 4/6
# P(W) = P(AN) * P(W assuming AN)   +   P(AC) * P(W assuming AC)  +  P(AT) * P(W assuming AT)
# P(W) = 15/36 * 1  +  15/36 * 0  +  6/36 * 4/6 = 19/36
```

```python
#concept of independent events
# fair coin rolling N times, total outcome => 2**N times
# FH - first event is Head
# FT - first event is Tail
# SH - second event is Head
# ST - second event is Tail
# P(SH) - 1/2 (if the second event is unaffected by the first event)
# P(SH assuming FH) = P(SH assuming FT) = 1/2 (if the second event is unaffected by the first event)
```

```python
#assume first event outcomes A and B and second event outcomes W and L
# if the second event outcomes does not depend on the first event,
# P(W) = P(W assuming A) = P(W assuming B) => first and second events are indepenent
# if two events are independent => prob of both A and W occur => P(A) * P(W)
```

```python
#P(L) : prob of left handed voters
#P(R) : prob of right handed voters
#P(N) : prob of candidate N
#P(C) : prob of candidate C
```

```python
#P(L) = 1/5
#P(R) = 4/5
#P(N) + P(C) = 1
#P(N) = P(N and L) + P(N and R)
#P(N and L) = P(N assuming L) * P(L) = 3/4 * 1/5 = 3/20
#P(N and R) = P(N assuming R) * P(R) = 2/5 * 4/5 = 8/25
#P(N) = 3/20 + 8/25 = 15/100 + 32/100 = 47/100
```

```python
#Bayes's Law
#P(L assuming N) is NOT always equal to P(N assuming L)
#In an extreme case, if L and N are completely independent events, they don't have any influence on each,
#P(N assuming L) = P(N) and P(L assuming N) = P(L)
#P(L assuming N) * P(N) = P(N assuming L) * P(L)
```

```python
#Notice that when N and L are completely independent, the probability of both N and L occurring, by the multiplication principle,
#P(N and L) = P(N) * P(L)
```

```python
#5.3 Practice Problem 3
#a test with 99 percent reliability => P(positive assuming sick) = 99/100
#that only one percent of the people in the country are sick => P(sick) = 1/100
#If the patient tests positive, what are the chances the patient is sick? => P(sick assuming positive)
#Using Bayes's Law
#P(A assuming B) * P(B) = P(B assuming A) * P(A)
# P(A assuming B) / P(A) = P(B assuming A) / P(B)
#P(positive assuming sick) * P(sick) = P(sick assuming positive) * P(positive)
#P(positive assuming sick) / P(positive) = P(sick assuming positive) / P(sick)
#P(positive) = P(positive and sick) + P(positive and well)
#P(positive and sick) = P(positive assuming sick) * P(sick) = 99/100 * 1/100 = 99/(10**4)
#P(postive and well) = P(postive assuming well) * P(well) = (1-99/100) * (1-1/100) = 99/(10**4)
#P(positive) = 99/(10**4) + 99/(10**4) = 198/(10**4)
#P(sick assuming positive) = (P(positive assuming sick) * P(sick)) / P(positive) = (99/100 * 1/100) / (198/(10**4)) = 99/198 = 1/2
```
